#### Model Building & Training

```python
# Defining the data and the target

X = data.drop(["quality"], axis = 1)


def categorize_quality(quality):
    if quality >= 7:
        return 1
    else:
        return 0

Y = data['quality'].apply(categorize_quality) #label binarization to create categories for the quality of wine; 1 = Good, 0 = Poor
```

```python
# Checking the binarization

Y.unique()
```

```python
# Split the data into train and test data with a test size of 20%

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```


Model Training


```python
# assigning the Random Forest Classifer to model

model = RandomForestClassifier()
```

```python
# Fitting the model on the train data

model.fit(X_train,Y_train)
```

```python
X_pred = model.predict(X_test)
```

```python
# Accuracy of model on the test data

accuracy = accuracy_score(X_pred,Y_test)
```
